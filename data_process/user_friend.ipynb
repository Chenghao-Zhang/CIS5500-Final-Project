```python
fri_kept_col=['user_id', 'friends']
fri_df = u_df.loc[:, fri_kept_col]
fri_df = fri_df.dropna(subset=['friends'])
fri_df['friends'] = fri_df['friends'].str.split(',')
fri_df = fri_df.explode('friends')
fri_df['friends'] = fri_df['friends'].str.strip()

pairs_df = pd.DataFrame({
    'user_a': np.where(fri_df['user_id'] < fri_df['friends'], fri_df['user_id'], fri_df['friends']),
    'user_b': np.where(fri_df['user_id'] > fri_df['friends'], fri_df['user_id'], fri_df['friends'])
})
pairs_df = pairs_df.drop_duplicates().dropna()
pairs_df = pairs_df[pairs_df['user_a'] != pairs_df['user_b']]
pairs_df.to_csv('/mnt/data/cleaned_befriend.csv', index=False)
pairs_df.to_csv('/content/drive/My Drive/550 project/cleaned_befriend', index=False, encoding='utf-8')
```
